## 3. Remove duplicated images

```python
# Remove duplicated images inside each class
def dhash(image, hash_size=8):
    return hash(str(image.tobytes()))
    
def find_duplicates(folder):
    duplicates = []
    hash_keys = {}
    for filename in os.listdir(folder):
        image_path = os.path.join(folder, filename)
        with Image.open(image_path) as img:
            h = dhash(img)
        if h in hash_keys:
            duplicates.append(image_path)
        else:
            hash_keys[h] = image_path
    return duplicates

def delete_duplicates(duplicates):
    if not duplicates:
        print("No duplicate files found.")
    else:
        for duplicate_path in duplicates:
            # Delete duplicate files
            os.remove(duplicate_path)
            print(f"Deleted duplicate: {duplicate_path}")

def process_folder(folder_path):
    duplicates = find_duplicates(folder_path)
    delete_duplicates(duplicates)

# Specify the paths
base_folder = 'C:/My_Files/UpWork_jobs/refrigerator_object_detection/refiragerator_dataset'
labels = ['apple', 'avocado', 'banana', 'beef_mince', 'beetroot', 'bellpepper', 'beverage', 'cabbage', 'carrot', 'cauliflower', 'chicken_thighs', 
          'cucumber', 'dairy', 'egg', 'eggplant', 'grapes', 'iceCream', 'jalepeno', 'lemon', 'mango', 'meat', 'mutton', 'orange', 'peas',
          'pineapple', 'steak', 'strawberry',  'tomato', 'watermelon']  # 29 classes/labels

# Process each folder for each label
for label in labels:
    folder_path = os.path.join(base_folder, label)
    process_folder(folder_path)
```

## 4. Resize and Convert from PNG/JPEG to JPG

```python
# Define the target image size to (224x224) RGB and convert to JPG only
target_size = (224, 224)
# Specify the source directory containing the images
source_dir = "C:/My_Files/UpWork_jobs/refrigerator_object_detection/refiragerator_dataset"
# Define the target directories
apple_dir = os.path.join(source_dir, "apple")
avocado_dir = os.path.join(source_dir, "avocado")
banana_dir = os.path.join(source_dir, "banana")
beef_mince_dir = os.path.join(source_dir, "beef_mince")
beetroot_dir = os.path.join(source_dir, "beetroot")
bellpepper_dir = os.path.join(source_dir, "bellpepper")
beverage_dir = os.path.join(source_dir, "beverage")
cabbage_dir = os.path.join(source_dir, "cabbage")
carrot_dir = os.path.join(source_dir, "carrot")
cauliflower_dir = os.path.join(source_dir, "cauliflower")
chicken_thighs_dir = os.path.join(source_dir, "chicken_thighs")
cucumber_dir = os.path.join(source_dir, "cucumber")
dairy_dir = os.path.join(source_dir, "dairy")
egg_dir = os.path.join(source_dir, "egg")
eggplant_dir = os.path.join(source_dir, "eggplant")
grapes_dir = os.path.join(source_dir, "grapes")
iceCream_dir = os.path.join(source_dir, "iceCream")
jalepeno_dir = os.path.join(source_dir, "jalepeno")
lemon_dir = os.path.join(source_dir, "lemon")
mango_dir = os.path.join(source_dir, "mango")
meat_dir = os.path.join(source_dir, "meat")
mutton_dir = os.path.join(source_dir, "mutton")
orange_dir = os.path.join(source_dir, "orange")
peas_dir = os.path.join(source_dir, "peas")
pineapple_dir = os.path.join(source_dir, "pineapple")
steak_dir = os.path.join(source_dir, "steak")
strawberry_dir = os.path.join(source_dir, "strawberry")
tomato_dir = os.path.join(source_dir, "tomato")
watermelon_dir = os.path.join(source_dir, "watermelon")

# Resize, save, and delete original JPEG files in the corresponding folders
for folder_name in [apple_dir, avocado_dir, banana_dir, beef_mince_dir, beetroot_dir, bellpepper_dir, beverage_dir, cabbage_dir, carrot_dir, cauliflower_dir, chicken_thighs, cucumber_dir, dairy_dir, egg_dir, eggplant_dir, 
                    grapes_dir, iceCream_dir, jalepeno_dir, lemon_dir, mango_dir, meat_dir, mutton_dir, orange_dir, peas_dir, pineapple_dir, steak_dir, strawberry_dir, tomato_dir, watermelon_dir]:
    for filename in os.listdir(folder_name):
        image_path = os.path.join(folder_name, filename)
        image = cv2.imread(image_path)
        resized_image = cv2.resize(image, target_size)
        # Rename the resized image to JPG format
        new_filename = os.path.splitext(filename)[0] + '.jpg'
        new_image_path = os.path.join(folder_name, new_filename)
        cv2.imwrite(new_image_path, resized_image)
        # Delete the original JPEG file
        if filename.lower().endswith(".jpeg"):
            os.remove(image_path)
        # Delete the original PNG file
        if filename.lower().endswith(".png"):
            os.remove(image_path)
```

## 5. Rename

```python
# Rename images
def rename_files_in_folders(root_folder):
    for folder_name in os.listdir(root_folder):
        folder_path = os.path.join(root_folder, folder_name)

        if os.path.isdir(folder_path):
            for filename in os.listdir(folder_path):
                if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
                    image_path = os.path.join(folder_path, filename)

                    # Extract filename without extension
                    filename_without_extension = os.path.splitext(filename)[0]

                    # Create new names
                    new_image_name = f"{folder_name}_{filename_without_extension}.jpg"

                    new_image_path = os.path.join(folder_path, new_image_name)

                    # Rename image file
                    os.rename(image_path, new_image_path)

# Replace 'your_root_folder' with the actual path to your root folder containing the 29 folders
root_folder = 'C:/My_Files/UpWork_jobs/refrigerator_object_detection/all_datasets/refiragerator_dataset'

# Call the function to rename files
rename_files_in_folders(root_folder)
```
